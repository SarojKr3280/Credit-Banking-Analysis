```python
# Ensure the date format is correct and extract month and year

spend_df['Month'] = pd.to_datetime(spend_df['Month'], format='%d-%b-%y')
spend_df['YearMonth'] = spend_df['Month'].dt.to_period('M')
spend_df
```

```python
# Ensure the date format is correct and extract month and year

repayment_df['Month'] = pd.to_datetime(repayment_df['Month'], format='%d-%b-%y')
repayment_df['YearMonth'] = repayment_df['Month'].dt.to_period('M')
repayment_df
```

```python
# monthly spend of each customer

monthlySpend = spend_df.groupby(['Customer','YearMonth'])['Amount'].sum().astype(int).reset_index()
print(type(monthlySpend))
monthlySpend
```

```python
# monthly repayment of each customer

monthlyRepayment = repayment_df.groupby(['Customer', 'YearMonth'])['Amount'].sum().astype(int).reset_index()
print(type(monthlyRepayment))
monthlyRepayment
```

```python
#  Highest paying customers.

hpc = monthlyRepayment.sort_values(['YearMonth', 'Amount'], ascending = [True, False])
hpc.style.hide()
```

```python
# Calculate the due amount 

due_amount_df = monthlySpend.merge(monthlyRepayment, on=['Customer', 'YearMonth'], how='left', suffixes=('_spend', '_repay'))
due_amount_df
due_amount_df['DueAmount'] = due_amount_df['Amount_spend'] - due_amount_df['Amount_repay'].fillna(0)
due_amount_df
```

```python
# Filter out rows where DueAmount is negative

due_amount_df = due_amount_df[due_amount_df['DueAmount'] > 0]
due_amount_df
```

```python
# Calculate interest on due amount

due_amount_df['Interest'] = due_amount_df['DueAmount'] * 0.029
due_amount_df
```

```python
# segment in which people are spending more

customer_segment = customer_df[['Customer', 'Segment']] 
spend = spend_df[['Customer', 'YearMonth', 'Amount']] 
customer_segment_spend = pd.merge(customer_segment, spend, on='Customer', how='left') 

# Calculated total spending per segment
segment_spend = customer_segment_spend.groupby([ 'YearMonth','Segment'])['Amount'].sum().astype(int).reset_index()
segment_spend.sort_values(['YearMonth','Amount'], ascending  = [True, False])
```

```python
Segment_spend_max = segment_spend.sort_values(['YearMonth','Amount'])
Segment_spend_month_max = Segment_spend_max.groupby('YearMonth').apply(lambda x: x[x['Amount'] == x['Amount'].max()]).style.hide()
Segment_spend_month_max
```

```python
segment_spend = customer_segment_spend.groupby('Segment')['Amount'].sum().astype(int).reset_index()
segment_spend
```

```python
customer_segment = customer_df[['Customer', 'Segment']] 
repayment = repayment_df[['Customer', 'YearMonth', 'Amount']] 
customer_segment_repayment = pd.merge(customer_segment, repayment, on='Customer', how='left') 

# Calculated total repayment per segment
segment_repayment = customer_segment_repayment.groupby([ 'Segment'])['Amount'].sum().astype(int).reset_index()
segment_repayment
```

```python
spend_repayment = pd.merge(segment_spend, segment_repayment, on='Segment', how='left', suffixes = ('_spent','_repaid'))
spend_repayment
spend_repayment['Profit'] = spend_repayment['Amount_spent'] - spend_repayment['Amount_repaid']
spend_repayment
```

```python
# most profitable segement

profit = spend_repayment[['Segment', 'Profit']]
profit
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.barplot(data = profit , x ='Segment', y ='Profit')
plt.title('Segment Wise Profit')
plt.xlabel('Segment')
plt.ylabel('Profit')
plt.xticks(rotation = 45)
plt.show()
```

```python
customer_age = customer_df_raw[['Customer', 'Age']] 
spend = monthlySpend[['Customer', 'YearMonth', 'Amount']] 
customer_age_spend = pd.merge(customer_age, spend, on='Customer', how='left') 
customer_age_spend
```

```python
ageWiseSpend = customer_age_spend.groupby(['Age','Customer'])['Amount'].sum().astype(int).reset_index()
ageWiseSpend
```

```python
for i in ageWiseSpend.index:
    if ageWiseSpend.loc[i, 'Age'] < 18: 
        ageWiseSpend.loc[i, 'Age'] = 'Infant'
    elif ageWiseSpend.loc[i, 'Age'] > 18 and ageWiseSpend.loc[i, 'Age'] < 30:
        ageWiseSpend.loc[i, 'Age'] = 'Youth'
    elif ageWiseSpend.loc[i, 'Age'] > 30 and ageWiseSpend.loc[i, 'Age'] < 60:
        ageWiseSpend.loc[i, 'Age'] = 'Adult'
    elif ageWiseSpend.loc[i, 'Age'] > 60:
        ageWiseSpend.loc[i, 'Age'] = 'Old'

ageWiseSpend
```

```python
ageGroupSpend = ageWiseSpend.groupby('Age')['Amount'].sum().astype(int).reset_index()
ageGroupSpend
```

```python
# Which age group is spending more money?

plt.figure(figsize=(10, 6))
sns.barplot(data=ageGroupSpend , x='Age', y='Amount')
plt.title('Total Spending by Age groups')
plt.xlabel('Age')
plt.ylabel('Total Spending Amount')
plt.xticks(rotation=45)
plt.show()
```
